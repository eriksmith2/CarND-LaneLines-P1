Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=10,):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """

    # separate left lane lines and right lane lines based on their slope, find average slopes and y-intercepts of lane lines

    left_slope = []
    right_slope = []
    left_b = []
    right_b = []

    for line in lines:
        for (x1, y1, x2, y2) in line:
            slope = (y2 - y1) / (x2 - x1)
            if -0.8 < slope < -0.4:
                left_slope.append(slope)
                left_b.append(y1 - slope * x1)
                left_b.append(y2 - slope * x2)
            elif 0.8 > slope > 0.4:
                right_slope.append(slope)
                right_b.append(y1 - slope * x1)
                right_b.append(y2 - slope * x2)

    avg_left_slope = np.mean(left_slope)
    avg_right_slope = np.mean(right_slope)
    avg_left_b = np.mean(left_b)
    avg_right_b = np.mean(right_b)

    # choose max and min y boundaries to draw avg lane lines between

    max_y = 540
    min_y = 325

    # draw lines

    cv2.line(img, (int((min_y - avg_left_b) / avg_left_slope), min_y),
             (int((max_y - avg_left_b) / avg_left_slope), max_y),
             color, thickness)
    cv2.line(img, (int((min_y - avg_right_b) / avg_right_slope),
             min_y), (int((max_y - avg_right_b) / avg_right_slope),
             max_y), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
def find_lanes(img):

    #define Hough variables
    rho = 1
    theta = np.pi/180
    threshold = 15
    min_line_length = 10
    max_line_gap = 5

    #Read image and convert to grayscale
    gray=grayscale(img)
    plt.imshow(gray, cmap='gray')
    plt.savefig('gray.png')

    #Apply gausian blur to grayscale image
    blur_gray=gaussian_blur(gray, 5)
    plt.imshow(blur_gray, cmap='gray')
    plt.savefig('blur_gray.png')

    #Perform canny edge detection. Thresholds are computed using Otsu’s Binarization as recommended here: https://carnd-udacity.atlassian.net/wiki/questions/22678863/why-arent-we-using-otsus-threshold
    high_threshold, _ = cv2.threshold(blur_gray, 0, 255, cv2.THRESH_OTSU)
    low_threshold = 0.5 * high_threshold
    edges=canny(blur_gray, low_threshold, high_threshold)
    plt.imshow(edges, cmap='Greys_r')
    plt.savefig('edges.png')
    
    #Refine image to only consider a region on interest
    imshape = edges.shape
    vertices = np.array([[(0,imshape[0]),(450, 325), (525, 325), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges=region_of_interest(edges, vertices)
    plt.imshow(masked_edges, cmap='Greys_r')
    plt.savefig('masked_edges.png')

    #Draw Hough lines over edges detected in canny image
    lines=hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    plt.imshow(lines)
    plt.savefig('lines.png')

    return weighted_img(lines, img, α=0.8, β=1., λ=0.)    
```

```python
#define and read in image to be processed
img='solidWhiteCurve.jpg'
img=mpimg.imread('test_images/'+img)

lanes=find_lanes(img)
plt.imshow(lanes)
plt.savefig('lanes.png')
```

## Reflections: 
My current pipeline follows these basic steps:
1. Read in image
2. Convert image to grayscale
3. Apply gausian blur to image to reduce noise
4. Perform Canny edge detection on image to isolate edges in the image
5. Isolate only the region that should contain the lane lines which is relatively constant for the test images and videos
6. Hough lines are created using the isolated edges
7. Divide the Hough lines into those that are likely the left lane line and those that are likely the right lane line. This is done by looking at the slopes of the lines - lines with a positive slope are likely part of the right lane line and lines with a negative slope are likely part of the left lane line.
8. The average slope and y-intercept of the left and right lane line is determined and an average line is created by inputting the y values for the bottom of the image and the approximate horizon, computing the corresponding x values of the average lines at those points, and drawing a line between the points.

As can be seen in the challenge video, my pipeline has issues when the road has a relatively large curvature that causes the isolated region where the lane lines are assumed to be to be incorrect. One way to solve this issue would be to have some sort of feedback where the shape of the isolated area is changed based on the lines detected in the previous frame.

My pipeline also has issues when there are shadows across the road that cause false positives in the Canny edge detection algorithm. There are likely elegant ways to solve this issue with better preprocessing of the grayscale image - I will need to do more research into ways to avoid ilumination problems.

Finally, my current pipeline is not as smooth as the example solution when drawing lines on each frame of the videos. This could be solved by adding a hysterisis to the lines being drawn. One way to do this would be to average the previous frame's line and the current frame's calculated line. This would help make the output a little smoother and could also help reduce the effect of outlier points on the output of the pipeline.
